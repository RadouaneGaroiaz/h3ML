#### Regression Liniere Simple

Faisons une prédiction sur le salaire d'un développeur basé sur leurs expériences.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X=np.array([0,3,6,8])
Y=np.array([35,45,65,80])
print(X,Y)
```

```python
plt.plot(X,Y,'*')
plt.xlabel("Années d'expérience | notre explicative variable 'x'" )
plt.ylabel("Salaire | Target variable 'y'")
plt.title("Scratter plot / Le complot de l'éparpilleur ")
plt.savefig("./intuitive_scatter.png")
```

```python
def reg_plot(x,y,m):
    plt.scatter(x,y,c='pink',label="nos data")
    # Tracé de la courbe de prédiction en utilisant le modèle 'm' pour prédire les valeurs correspondantes à 'x'
    #x.reshape permet de remodeler un tableau
    # La courbe est tracée en bleu ('blue') et une légende "prediction de la courbe" est ajoutée
    plt.plot(x, m.predict(x.reshape(-1, 1)), color='blue',label="prediction de la courbe")
    plt.xlabel("explicative variable 'x' ")
    plt.ylabel("target variable 'y'")
    plt.legend()
    return None
```

```python
# Importation de la classe LinearRegression depuis la bibliothèque de modèles linéaires de scikit-learn
linear_model = LinearRegression()

# Entraînement du modèle linéaire avec les données X et Y.
# X.reshape(-1, 1) est utilisé pour remodeler le tableau X en une matrice unidimensionnelle
# ayant une seule colonne, ce qui est attendu par la méthode fit.
linear_model.fit(X.reshape(-1, 1), Y)

# Appel à la fonction reg_plot pour tracer un nuage de points et une courbe de prédiction basée sur le modèle linéaire
reg_plot(X, Y, linear_model)

plt.savefig("./approche_intuitive.png")
```

### Prédire le salaire par rapport à l'expérience

```python
import pandas as pan
df = pan.read_csv("data/salary_Data.csv", sep=',')
df.head()
```

```python
df=df[["YearsExperience","Salary"]]
#Pour afficher juste les deux colonnes qui nous intérésse
```

```python
X=df.YearsExperience
Y=df.Salary

print(X,Y)

plt.plot(X,Y,'*')
plt.xlabel("Année d'expériences |  our explicative variable 'x' ")
plt.ylabel("Salaire | Target variable 'y'")
plt.title("Année d'expériences vs Salaire")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

linear_model = LinearRegression()
linear_model.fit(np.array(X).reshape(-1, 1),np.array(Y))
```

```python
reg_plot(np.array(X),np.array(Y),linear_model)
```

```python
# Utilise le modèle linéaire (linear_model) pour prédire les valeurs cibles (Y_pred) en utilisant les données d'entrée X.
# np.array(X).reshape(-1, 1) redimensionne les données X en une matrice unidimensionnelle (une seule colonne),
# ce qui est attendu par la méthode predict du modèle linéaire.
Y_pred = linear_model.predict(np.array(X).reshape(-1, 1))

# Calcule l'erreur quadratique moyenne (MSE) entre les valeurs cibles réelles (Y) et les valeurs prédites (Y_pred).
# La valeur est affichée avec une précision de 2 décimales en utilisant le format %.2f.
print("Erreur quadratique moyenne: %.2f"
      % mean_squared_error(Y, Y_pred))

# Calcule la racine carrée de l'erreur quadratique moyenne (RMSE) entre les valeurs cibles réelles (Y) et les valeurs prédites (Y_pred).
# La valeur est affichée avec une précision de 2 décimales en utilisant le format %.2f.
print("Racine carrée de l'erreur quadratique moyenne: %.2f"
      % np.sqrt(mean_squared_error(Y, Y_pred)))

# Calcule le coefficient de détermination (R carré) entre les valeurs cibles réelles (Y) et les valeurs prédites (Y_pred).
# La valeur est affichée avec une précision de 2 décimales en utilisant le format %.2f.
print("R carré: %.2f"
      % r2_score(Y, Y_pred))
```
